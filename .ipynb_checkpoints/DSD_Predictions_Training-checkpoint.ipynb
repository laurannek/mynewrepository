# Verification of the DSD Model

  ## 1. Verify which Dropzones have no activity 
     -(When Dropzone D and/or S is equal to 0)

  ## 2. Verify which Dropzones that had activity and have no activity anymore
     -(When Dropzone had a positive D and/or S - in the last prediction-, check that it is not 0 in the next models)

# 1. Dropzones that have no activity 
### (For each Org, over weeks)

```python
from google.cloud import storage
import pickle

# get GS bucket and setup token
gs = storage.Client("fleets-production")
bucket = gs.get_bucket("noa-kraken-production")

# get list of DSD files
DSDlist = list(bucket.list_blobs(prefix='dropzone_supply_demand/models/baseline/dt_None_tf_10T_tw_70'))

# Iterate on each prediction file, get Dropzones that have no Demand neither Supply predictions                                 
a=0
while a<len(DSDlist):
    for Org in pickle.loads(DSDlist[a].download_as_string())["balance"].keys():
        print(Org)
        a=a+1
        for DSD_el in DSDlist:
                print(DSD_el.path)
                dsdlist = pickle.loads(DSD_el.download_as_string())
                supply_dz_df = dsdlist["balance"][Org]["median"].max().reset_index()
                demand_dz_df= dsdlist["balance"][Org]["median"].min().reset_index()
                null_supply= supply_dz_df[supply_dz_df[0] == 0]['dropzone']
                null_demand= demand_dz_df[demand_dz_df[0] == 0]['dropzone']
                print(set(null_supply).intersection(set(null_demand)))
```

# 2. Dropzones that had activity and do not have activity anymore
### (for each Org, over weeks)

```python
from google.cloud import storage
import pickle

# get GS bucket and setup token
gs = storage.Client("fleets-production")
bucket = gs.get_bucket("noa-kraken-production")

#import numpy
import numpy as np


# get list of DSD files
DSDlist = list(bucket.list_blobs(prefix='dropzone_supply_demand/models/baseline/dt_None_tf_10T_tw_70'))

# Iterate on each prediction file, get Dropzones that have positive Demand and Supply predictions
a=0
while a<len(DSDlist):
    for Org in pickle.loads(DSDlist[a].download_as_string())["balance"].keys():
        print(Org)
        a = a+1
        positive_ds = set()
        for DSD_el in DSDlist:
            print(DSD_el.path)
            dsdlist = pickle.loads(DSD_el.download_as_string())
            supply_dz_df = dsdlist["balance"][Org]["median"].max().reset_index()
            demand_dz_df = dsdlist["balance"][Org]["median"].min().reset_index()
            positive_supply = supply_dz_df[supply_dz_df[0] > 0]["dropzone"]
            positive_demand = demand_dz_df[demand_dz_df[0] < 0]["dropzone"]
            
            disappearing_ds = positive_ds.difference(set(positive_supply).union(set(positive_demand)))
            #print(positive_ds) - to verify that positive_ds is the list of drop with positive ds in the previous iteration
            #print(set(positive_supply).union(set(positive_demand))) - to verify the list of drop with positive ds in this iteration
            print(disappearing_ds)
            positive_ds = set(positive_supply).union(set(positive_demand))
    
```
